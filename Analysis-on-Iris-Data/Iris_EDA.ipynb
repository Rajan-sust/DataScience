```
import seaborn as sns
import pandas as pd
```

```
source = 'https://raw.githubusercontent.com/Rajan-sust/ML/master/Dataset/iris/trian-test.csv'
data = pd.read_csv(source, names=['sepal-length','sepal-width', 'petal-length','petal-width','class'])
```

```
data.head()
```

```
data.describe()
```

```
sns.set() 
sns.lmplot(x = 'sepal-length', y = 'petal-length', data = data, hue = 'class', palette='Set2')
```

```
sns.lmplot(x = 'sepal-length', y = 'petal-length', data = data, hue = 'class', palette='Set2', fit_reg = False)
```

```
sns.pairplot(data = data, hue = 'class', palette = 'Set2')
```

```
sns.boxplot(data = data, orient = 'h')
```
